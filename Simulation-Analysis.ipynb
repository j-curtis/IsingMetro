```python
import numpy as np
from matplotlib import pyplot as plt 
import Simulate as sim
import Metropolis1D as m1d
import Metropolis2D as m2d
```

```python
"""Correlation time for 1-D and no external field"""
"""See HW8 Problem 2"""
"""We create an ensemble of simulations and store the observables"""
d = 1
h = 0.0
J = 0.5 
L = 500
nSteps = 1000*L
nTrials = 3

ensemble = sim.Ensemble(d,L,J,h,nSteps,nTrials)
ensemble.generate()
ensemble.avg()
```

```python
"""Energy vs. Time"""
plt.plot(ensemble.energy[0],color="red")

plt.ylim(-1.0,0.0)
plt.xlim(0,nSteps)

plt.title("Energy per Site vs. Time")
plt.xlabel("t (steps)")
plt.ylabel(r"Energy Density ($k_B T$)")
plt.text(nSteps/2,-.6,"L=%s\nJ=%s\nh=%s"%(L,J,h))

#plt.savefig("../Plots/1D/two-runs_energy_v_time.pdf")
plt.show()
```

```python
"""Mean Energy vs. Time"""
plt.plot(ensemble.energyMean,color="blue")

plt.ylim(-1.0,0.0)
plt.xlim(0,nSteps)

plt.title("Mean Energy per Site vs. Time")
plt.xlabel("t (steps)")
plt.ylabel(r"Energy Density ($k_B T$)")
plt.text(nSteps/2,-.6,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

#plt.savefig("../Plots/1D/mean_energy_v_time.pdf")
plt.show()
```

```python
"""Mean Magnetization vs. Time"""
plt.plot(ensemble.magMean,color="blue")
plt.axhline(color="gray")

plt.ylim(-1.0,1.0)
plt.xlim(0,nSteps)

plt.title("Magnetization Density vs. Time")
plt.xlabel("Time (steps)")
plt.ylabel(r"Magnetization Density (unitless)")
plt.text(nSteps/2,.4,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

#plt.savefig("../Plots/1D/mean_mag_v_time.pdf")
plt.show()
```

```python
"""Mean Nearest-Neighbor Correlation vs. Time"""
plt.plot(ensemble.pairMean,color="blue")
plt.plot(ensemble.pair[0],color="red")

plt.ylim(-1,1)
plt.xlim(0,nSteps)

plt.title("NN Pair Correlation per Site vs. Time")
plt.xlabel("Time (steps)")
plt.ylabel("Correlation/L (unitless)")
plt.text(nSteps/2,-0.5,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

plt.show()
```

```python
"""Magnetization and Energy for Different Parameters in 1D"""
"""See HW8 Problem 4"""
"""We run a simulation and sample to generate an ensemble distribution"""
d = 1
L = 150
nSteps = 1000*L

jMin = 0.0
jMax = 2.0
numJ = 20

jList = np.linspace(jMin,jMax,numJ)

hMin = 0.0
hMax = 2.0 
numH = 20

hList = np.linspace(hMin,hMax,numH)

magMean = np.zeros(shape=(numJ,numH))
energyMean = np.zeros(shape=(numJ,numH))
```

```python
for j in np.arange(numJ):
    for h in np.arange(numH):
        run = sim.Simulation(d,L,jList[j],hList[h],nSteps)
        run.simulate()
        magMean[j,h] = run.meanMag
        energyMean[j,h] = run.meanEnergy
```

```python
plt.plot(run.mag)
plt.show()
```
