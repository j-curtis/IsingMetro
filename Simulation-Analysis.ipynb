```python
import numpy as np
from matplotlib import pyplot as plt 
import Simulate as sim
import Metropolis1D as m1d
import Metropolis2D as m2d
```

```python
"""Correlation time for 1-D and no external field"""
"""See HW8 Problem 2"""
"""We create an ensemble of simulations and store the observables"""
d = 1
h = 0.0
J = 0.5 
L = 250
nSteps = 50*L
nTrials = 500

ensemble = sim.Ensemble(d,L,J,h,nSteps,nTrials)
ensemble.generate()
ensemble.avg()
```

```python
"""Energy vs. Time"""
plt.plot(ensemble.energy[0],color="red")
plt.plot(ensemble.energy[1],color="green")

plt.ylim(-1.0,0.0)
plt.xlim(0,nSteps)

plt.title("Energy per Site vs. Time")
plt.xlabel("t (steps)")
plt.ylabel(r"Energy Density ($k_B T$)")
plt.text(nSteps/2,-.6,"L=%s\nJ=%s\nh=%s"%(L,J,h))

plt.show()
plt.savefig()
```

```python
"""Mean Energy vs. Time"""
plt.plot(ensemble.energyMean,color="blue")
plt.plot(ensemble.energy[0],color="red")

plt.ylim(-1.0,0.0)
plt.xlim(0,nSteps)

plt.title("Energy per Site vs. Time")
plt.xlabel("t (steps)")
plt.ylabel("E/L (unitless)")
plt.text(nSteps/2,-.6,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

plt.show()
```

```python
"""Mean Magnetization vs. Time"""
plt.plot(ensemble.magMean,color="blue")
plt.axhline(color="gray")
plt.plot(ensemble.mag[0],color="red")

plt.ylim(-1.0,1.0)
plt.xlim(0,nSteps)

plt.title("Magnetization per Site vs. Time")
plt.xlabel("Time (steps)")
plt.ylabel("M/L (unitless)")
plt.text(nSteps/2,.4,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

plt.show()
```

```python
"""Mean Nearest-Neighbor Correlation vs. Time"""
plt.plot(ensemble.pairMean,color="blue")
plt.plot(ensemble.pair[0],color="red")

plt.ylim(-1,1)
plt.xlim(0,nSteps)

plt.title("NN Pair Correlation per Site vs. Time")
plt.xlabel("Time (steps)")
plt.ylabel("Correlation/L (unitless)")
plt.text(nSteps/2,-0.5,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

plt.show()
```

```python
"""Energy Correlations"""
plt.imshow(ensemble.energyCorr)

plt.title("Normalized Energy-Energy Correlation")
plt.xlabel(r"$t_1$ (steps)")
plt.ylabel(r"$t_2$ (steps)")
plt.colorbar()

plt.text(nSteps*1.5,nSteps/2,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

plt.show()
```
