```python
import numpy as np
from matplotlib import pyplot as plt 
import Simulate as sim
import Metropolis1D as m1d
import Metropolis2D as m2d
```

```python
"""Correlation time for 1-D and no external field"""
"""See HW8 Problem 2"""
"""We create an ensemble of simulations and store the observables"""
d = 2
h = 0.0
J = .2 
L = 50
nSteps = 500*L
nTrials = 10

ensemble = sim.Ensemble(d,L,J,h,nSteps,nTrials)
ensemble.generate()
ensemble.avg()
```

```python
"""Mean Magnetization vs. Time"""
plt.plot(ensemble.magMean,color="red")
#plt.axhline(color="gray")

plt.xlim(0,nSteps)

plt.title("2D Magnetization Density vs. Time")
plt.xlabel("Time (steps)")
plt.ylabel(r"Magnetization Density (unitless)")
plt.text(nSteps/3,.9996,"L=%s\nJ=%s\nh=%s\n\n%s Trials"%(L,J,h,nTrials))

#plt.savefig("../Plots/2D/mean_mag_v_time_2.pdf")
plt.show()
```

```python
"""Magnetization and Energy for Different Parameters in 1D"""
"""See HW8 Problem 4"""
"""We run a simulation and sample to generate an ensemble distribution"""
d = 1
L = 150
nSteps = 1000*L

jMin = 0.0
jMax = 2.0
numJ = 10

jList = np.linspace(jMin,jMax,numJ)

hMin = 0.0
hMax = 2.0 
numH = 20

hList = np.linspace(hMin,hMax,numH)

magMean = np.zeros(shape=(numJ,numH))
energyMean = np.zeros(shape=(numJ,numH))
```

```python
for j in np.arange(numJ):
    for h in np.arange(numH):
        run = sim.Simulation(d,L,jList[j],hList[h],nSteps)
        run.simulate()
        run.timeAverage(10*L,4*L)
        magMean[j,h] = run.meanMag
        energyMean[j,h] = run.meanEnergy
```

```python
plt.imshow(magMean,origin='lower')
plt.colorbar()

plt.title("Magnetization vs. Field and Coupling")
plt.xlabel(r"h ($k_B T$)")
plt.ylabel(r"J ($k_B T$)")

plt.text(12,2,"L=%s"%(L))

plt.savefig("../Plots/1D/mag_v_h_J.pdf")
plt.show()
```

```python
plt.imshow(energyMean,origin='lower')
plt.colorbar()

plt.title("Internal Energy vs. Field and Coupling")
plt.xlabel(r"h ($k_B T$)")
plt.ylabel(r"J ($k_B T$)")

plt.text(12,2,"L=%s"%(L))

plt.savefig("../Plots/1D/E_v_h_J.pdf")
plt.show()
```
