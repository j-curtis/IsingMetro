```python
import numpy as np
import matplotlib.pyplot as plt
import Metropolis2D as m2d
```

```python
"""Parameters"""
L = 100 #Lattice Size
J = 0.35 #Spin Coupling 
h = 0.0 #External Field 
nSteps = 1000*1000 #The number of steps we use to simulate the lattice 
```

```python
latt = m2d.Lattice(L,J,h)
```

```python
"""Plot of Lattice State"""
plt.imshow(latt.spins)

plt.xlim(0,L)
plt.ylim(0,L)

plt.title("Lattice State")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
for i in np.arange(nSteps):
    latt.metropolisStep()
```

```python
"""Plot of Lattice State"""
plt.imshow(latt.spins)

plt.xlim(0,L)
plt.ylim(0,L)

plt.title("Lattice State")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
